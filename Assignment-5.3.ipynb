```python
import numpy as np
import pandas as pd

import json
```

```python
DataSet = pd.read_csv("https://raw.githubusercontent.com/pcsanwald/kaggle-titanic/master/train.csv",sep=',')
DataSet.shape
DataSet
```

<h1>Convertig dataframe to Json</h1>

```python
DsToJs = DataSet.to_json(orient='records')
```

```python
DsToJs
```

<h1>#2 Read the JSON object and convert it to pandas data frame</h1>

```python
dict = json.loads(DsToJs);
df = pd.DataFrame.from_dict(dict, orient='columns')
df.head()
```
